Luego de varias exploracione obtenemos el Dataset final que se va a procesar.

```python
credit = pd.read_csv('default of credit card clients.csv', header = 1)
credit = credit.drop(columns = ['ID', 'SEX', 'MARRIAGE', 'AGE', 'LIMIT_BAL', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'])
indexAges = credit[ credit['EDUCATION'] == 0].index
credit.drop(indexAges, inplace=True)
indexAges = credit[ credit['EDUCATION'] == 4].index
credit.drop(indexAges, inplace=True)
indexAges = credit[ credit['EDUCATION'] == 5].index
credit.drop(indexAges, inplace=True)
indexAges = credit[ credit['EDUCATION'] == 6].index
credit.drop(indexAges, inplace=True)
corrMat = credit.corr()
print(corrMat)
```
